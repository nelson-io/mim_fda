1.1.b

```python
def sqroot_2(n,m):
  x1 = int(n ** (1/2))
  x2 = int(m ** (1/2))
  return [x1,x2]

print(sqroot_2(10,20))
```

1.1.c

```python
def sqroot_n(*args):
  x = [int(arg ** (1/2)) for arg in args]
  return x

print(sqroot_n(1,2,3,10,20,30))
```

```python
def prod(n,m): 
  total = 0
  counter = 0
  while counter < m:
    total += n
    counter += 1
  return total

print(prod(10,5))
```

```python
def pot(m,n):
    if n==0:
        return 1
          
    answer = m
    increment= m
      
    for i in range(1,n):
        for j in range(1,m):
            answer+=increment
        increment=answer
    return answer
  

print(pot(5,3))

```

```python
def product_n(*args):
  x = 1
  for i in args:
    x *= i
  return x
print(product_n(5,5,5))
```

```python
def factorial(n):
  x = range(1,n+1)
  out = product_n(*x)
  return out

print(factorial(4))
```

```python
def combin(n,k):
  if k > n:
    return None
  else:
    x = factorial(n)/(factorial(k)*factorial(n-k))
    return x

print(combin(5,10))
print(combin(10,2))

```

```python
def princombin(n):
  for i in range(0,n+1):
    print(combin(n, i))


princombin(10)
```
