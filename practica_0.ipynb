```python
def divisors(x):
  x = [i for i in range(1,x) if x % i == 0]
  return x

```

```python
def friendly_nums_qty(start = 1, end = 100):
  fnlist = []
  for i in range(start,end):
    val = sum(divisors(sum(divisors(i))))
    if val == i:
      fnlist.append(val)
  return len(fnlist)
```

```python
print(friendly_nums_qty(1,100))
print(friendly_nums_qty(1,5500))
print(friendly_nums_qty(10000,20000))


```

```python
def filtrar_palabras_mas_largas(palabras):
  if palabras == []:
    return []
  else:
    max_length = max([len(word) for word in palabras])
    x = [word for word in palabras if len(word) == max_length]
    return x
	


lista_vacia = []
lista_palabras_1 = ["casa"]
lista_palabras_2 = ["casa", "eso", "pez", "la"]
lista_palabras_3 = ["casa", "eso", "pez", "la", "mero"]
lista_palabras_4 = ["serenidad", "pitaya", "delicia", "permitidos", "saltamontes"]
lista_palabras_5 = ["serenidad", "pitaya", "delicia", "permitidos", "saltamontes", "capricornio"]

print(filtrar_palabras_mas_largas(lista_vacia))
print(filtrar_palabras_mas_largas(lista_palabras_1))
print(filtrar_palabras_mas_largas(lista_palabras_2))
print(filtrar_palabras_mas_largas(lista_palabras_3))
print(filtrar_palabras_mas_largas(lista_palabras_4))
print(filtrar_palabras_mas_largas(lista_palabras_5))
```
